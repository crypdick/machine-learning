### Implementation: Define a Performance Metric
It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. For this project, you will be calculating the [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination), R<sup>2</sup>, to quantify your model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the *mean* of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**. _A model can be given a negative R<sup>2</sup> as well, which indicates that the model is **arbitrarily worse** than one that always predicts the mean of the target variable._

For the `performance_metric` function in the code cell below, you will need to implement the following:
- Use `r2_score` from `sklearn.metrics` to perform a performance calculation between `y_true` and `y_predict`.
- Assign the performance score to the `score` variable.

```python
# TODO: Import 'r2_score'
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

**Answer: ** The grid search takes a list of functions you want to test in your algorithm, and iteratively tests and scores each one. For each given model, you can specify the bounds for each individual parameter, and the grid search will test explore the parameter space to see which scores best. The grid search technique can be used if you want to brute force search all of your options for the best performing model/parameter combination. ### Question 8 - Cross-Validation
*What is the k-fold cross-validation training technique? What benefit does this technique provide for grid search when optimizing a model?*  
**Hint:** Much like the reasoning behind having a testing set, what could go wrong with using grid search without a cross-validated set?

```python
import numpy as np
from sklearn import cross_validation
from sklearn import datasets
from sklearn import svm

iris = datasets.load_iris()

# Split the iris data into train/test data sets with 30% reserved for testing
X_train1, X_test1, y_train1, y_test1 = cross_validation.train_test_split(iris.data, iris.target, test_size=0.3, random_state=0)

# Build an SVC model for predicting iris classifications using training data
clf1 = svm.SVC(kernel='linear', C=1, probability=True).fit(X_train1, y_train1)

# Now measure its performance with the test data with single subset
print clf1.score(X_test1, y_test1) 

# We give cross_val_score a model, the entire data set and its "real" values, and the number of folds:
scores1 = cross_validation.cross_val_score(clf1, iris.data, iris.target, cv=5)

# Print the accuracy for each fold:
print scores1

# And the mean accuracy of all 5 folds:
print scores1.mean()
```

**Answer: ** The K-fold cross validation is an algorithm which chunks your data into segments. Then, you reserve one of those chunks as a validation set, with the rest acting as a training set for your model. You train the model and score it. Then, you repeat the process, choosing a different chunk as the validation set. You repeat the process until each chunk has been used as a validation set exactly once. Then, you can use the summary statistics of the scores (especially mean score and variance) to benchmark the model.

This greatly increases the amount of training time, since you have to retrain the model on every conceavable combination of data chunks. If combined with grid search, it's repeating this process for every conceivable model parameters. However, you benefit from having models which have been trained and validated with all the data available to you. This is very powerful when using grid search because not only are you searching through the entire solution space, but you are getting summary statistics (especially mean and variance) for each model in the solution space. This helps find a model tailored to your needs, such as minimizing the bias-variance trade-off.

Each fold can be thought of as a trial in an experiment. More trials take more work, but they increase the confidence of the results. It prevents systematic errors occuring from not properly randomizing data before splitting it into test/training sets. It also avoids using the optimal parameters for a particular validation set, which might not be the globally optimum in the parameter space. It also lets you know how rebust your model is, since it is repeatedly being tested against a fresh validation set. The average score and score volatility lets you estimate the generalization error.

  - Assign this object to the `'regressor'` variable.
- Create a dictionary for `'max_depth'` with the values from 1 to 10, and assign this to the `'params'` variable.
- Use [`make_scorer`](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.make_scorer.html) from `sklearn.metrics` to create a scoring function object.
  - Pass the `performance_metric` function as a parameter to the object.
  - Assign this scoring function to the `'scoring_fnc'` variable.
- Use [`GridSearchCV`](http://scikit-learn.org/0.17/modules/generated/sklearn.grid_search.GridSearchCV.html) from `sklearn.grid_search` to create a grid search object.
  - Pass the variables `'regressor'`, `'params'`, `'scoring_fnc'`, and `'cv_sets'` as parameters to the object. 
  - Assign the `GridSearchCV` object to the `'grid'` variable.

```python
# TODO: Import 'make_scorer', 'DecisionTreeRegressor', and 'GridSearchCV'
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import make_scorer
from sklearn.grid_search import GridSearchCV  # spits a depreciation warning
from sklearn.cross_validation import ShuffleSplit

def fit_model(X, y):
    """ Performs grid search over the 'max_depth' parameter for a 
        decision tree regressor trained on the input data [X, y]. """
    # Create cross-validation sets from the training data
    
    #####################
    # there was a an error with the original code:
    # _init__() got multiple values for keyword argument 'n_splits'
    # I renamed n_splits to n_iter
    cv_sets = ShuffleSplit(X.shape[0], n_iter = 10, test_size = 0.2)

    # TODO: Create a decision tree regressor object
    regressor = DecisionTreeRegressor(random_state=0)

    # TODO: Create a dictionary for the parameter 'max_depth' with a range from 1 to 10
    params = {'max_depth': range(1,11,1)}

    # TODO: Transform 'performance_metric' into a scoring function using 'make_scorer' 
    scoring_fnc = make_scorer(performance_metric)

    # TODO: Create the grid search object
    grid = GridSearchCV(regressor, params, scoring=scoring_fnc, cv=cv_sets)

    # Fit the grid search object to the data to compute the optimal model
    grid = grid.fit(X, y)
    
    ##### adding this so that I can figure out why my answer was wrong
    #results = grid.grid_scores_

    # Return the optimal model after fitting the data
    return grid.best_estimator_#, results
```
